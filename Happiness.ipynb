# Data Cleaning

Encoding

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df['country']=LE.fit_transform(df['country'])
df['region']=LE.fit_transform(df['region'])
```

Removing the outliers

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
```

```python
threshold=3
print(np.where(z>3))
```

```python
df_new=df[(z<3).all(axis=1)]
```

Removed 9 rows.

Removal of skewness

```python
df_new.skew()
```

```python
df_new['standard_error']=np.log(df_new['standard_error'])
df_new['standard_error'].skew()
```

```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson',standardize=True)
c=['family']
df_new['family']=pt.fit_transform(df_new[['family']])
df_new['family'].skew()
```

```python
pt=PowerTransformer(method='yeo-johnson',standardize=True)
c=['health_(life_expectancy)']
df_new['health_(life_expectancy)']=pt.fit_transform(df_new[['health_(life_expectancy)']])
df_new['health_(life_expectancy)'].skew()
```

```python
pt=PowerTransformer(method='yeo-johnson',standardize=True)
c=['trust_(government_corruption)']
df_new['trust_(government_corruption)']=pt.fit_transform(df_new[['trust_(government_corruption)']])
df_new['trust_(government_corruption)'].skew()
```

```python
pt=PowerTransformer(method='yeo-johnson',standardize=True)
c=['generosity']
df_new['generosity']=pt.fit_transform(df_new[['generosity']])
df_new['generosity'].skew()
```

```python
df_new.skew()
```

Skewness of all the features is removed.

```python
lm.score(x_train,y_train)
```

```python
pred=lm.predict(x_test)
print("Predicted Score:",pred)
print("Actual Score:",y_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print("Error:")
print("Mean Absolute Error:",mean_absolute_error(y_test,pred))
print("Mean Squared Error:",mean_squared_error(y_test,pred))
print("Root Mean Squared Error:",np.sqrt(mean_squared_error(y_test,pred)))
```
